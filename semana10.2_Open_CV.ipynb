# Dibujar con el rat√≥n en una imagen

```python

# Definimos la funci√≥n para dibujar con el rat√≥n
def dibujar(event, x, y, etiquetas, parametros):
    if event == cv2.EVENT_LBUTTONDOWN:  # Cada vez que se presione el bot√≥n izquierdo del rat√≥n
        # Dibuja un c√≠rculo en la posici√≥n (x, y) con radio 50 y color azul
        cv2.circle(imagen8, (x, y), 50, (255, 0, 0), -1)

# Conectamos la imagen con la funci√≥n dibujar
cv2.namedWindow('mi_imagen8')  # Crea una ventana llamada 'mi_imagen8'
cv2.setMouseCallback('mi_imagen8', dibujar)  # Asocia la funci√≥n 'dibujar' con los eventos del rat√≥n en la ventana 'mi_imagen8'

# Creamos una imagen en negro donde vamos a pintar
imagen8 = np.zeros((500, 500, 3), np.uint8)

# Bucle principal para mostrar la imagen y detectar eventos del rat√≥n
while True:
    cv2.imshow('mi_imagen8', imagen8)  # Muestra la imagen en la ventana
    if cv2.waitKey(10) & 0xFF == 27:  # Espera 10 ms y verifica si se ha presionado la tecla 'Esc' (c√≥digo ASCII 27) para cerrar
        break
    if cv2.getWindowProperty('mi_imagen8', cv2.WND_PROP_VISIBLE) < 1:  # Verifica si la ventana ha sido cerrada
        break

# Destruye todas las ventanas creadas por OpenCV
cv2.destroyAllWindows()

```

# Dibujar arrastrando el rat√≥n sobre una imagen

```python

# Variables globales para almacenar el estado del dibujo y las coordenadas
dibujando = False
valorx = 0
valory = 0

# Definimos la funci√≥n para dibujar
def dibujar(event, x, y, etiquetas, parametros):
    global dibujando, valorx, valory
    # Cuando se presiona el bot√≥n izquierdo del rat√≥n
    if event == cv2.EVENT_LBUTTONDOWN:
        dibujando = True  # Comienza a dibujar
        valorx = x  # Guarda la coordenada x inicial
        valory = y  # Guarda la coordenada y inicial
    # Mientras se mueve el rat√≥n con el bot√≥n izquierdo presionado
    elif event == cv2.EVENT_MOUSEMOVE:
        if dibujando == True:
            # Dibuja un rect√°ngulo desde el punto inicial hasta el punto actual
            cv2.rectangle(imagen8, (valorx, valory), (x, y), (255, 0, 0), -1)
    # Cuando se suelta el bot√≥n izquierdo del rat√≥n
    elif event == cv2.EVENT_LBUTTONUP:
        dibujando = False  # Termina de dibujar
        # Dibuja el rect√°ngulo final
        cv2.rectangle(imagen8, (valorx, valory), (x, y), (255, 0, 0), -1)

# Conectamos la ventana de la imagen con la funci√≥n dibujar
cv2.namedWindow('mi_imagen9')
cv2.setMouseCallback('mi_imagen9', dibujar)

# Creamos una imagen en negro donde vamos a pintar
imagen8 = np.zeros((500, 500, 3), np.uint8)

# Bucle principal para mostrar la imagen y detectar eventos del rat√≥n
while True:
    cv2.imshow('mi_imagen9', imagen8)
    # Espera 10 ms y verifica si se ha presionado la tecla 'Esc' (c√≥digo ASCII 27) para cerrar
    if cv2.waitKey(10) & 0xFF == 27:
        break
    # Verifica si la ventana ha sido cerrada
    if cv2.getWindowProperty('mi_imagen9', cv2.WND_PROP_VISIBLE) < 1:
        break

# Destruye todas las ventanas creadas por OpenCV
cv2.destroyAllWindows()
```
