```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv(r'G:\OneDrive - Universidad Tecnologica del Peru\Estudios\Cursos\Analisis de Datos\Ciencia de Datos con Python IEEE AESS\codigo\base de datos\ejemplo1.csv') 
df.head() 
```

```python
df.count()
```

```python
df=df.drop(['Lanzamiento'], axis=1)
df.head()
```

```python
X_data = df.drop('Éxito', axis=1) #eliminamos la variable exito
y_exito = df['Éxito'] #variable objetivo
```

```python
X_data.head()
```

```python
y_exito.head()
```

```python
# Dividir los datos en conjuntos de entrenamiento (80%) y prueba (20%)
X_train, X_test, y_train, y_test = train_test_split(X_data, y_exito, test_size=0.3, random_state=42)
```

```python
# Crear y entrenar el modelo
model = LogisticRegression()  # Crear un modelo de regresión logística
model.fit(X_train, y_train)   # Entrenar el modelo con los datos de entrenamiento
```

```python
# Realizar predicciones
y_pred = model.predict(X_test)  # Predecir los resultados para los datos de prueba
```

```python
#  Evaluar el modelo
# Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
```

```python
# Reporte de clasificación con métricas adicionales
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Visualizar la matriz de confusión
# Heatmap con Seaborn
plt.figure(figsize=(5, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Visualizar la matriz de confusión
# Heatmap con Seaborn
plt.figure(figsize=(5, 5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot= True, fmt='g', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
