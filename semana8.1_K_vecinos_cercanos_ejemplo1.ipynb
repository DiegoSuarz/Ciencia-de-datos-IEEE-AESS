```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv(r'G:\OneDrive - Universidad Tecnologica del Peru\Estudios\Cursos\Analisis de Datos\Ciencia de Datos con PythonIEEE AESS\codigo\base de datos\ejemplo1_KNN.csv')
```

```python
df.head()
```

```python
X_data= df.drop([ 'Incumplimiento'], axis=1) 
X_data.head()
```

```python
y_incumpli= df['Incumplimiento'] #variable objetivo
y_incumpli.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_incumpli, test_size= 0.3, random_state = 42)
```

# K = 1

```python
knnmodelk1 = KNeighborsClassifier(n_neighbors = 1) #K == 1
knnmodelk1.fit(X_train, y_train)
```

```python
y_predict_k1= knnmodelk1.predict(X_test)
```

```python
cmk1= confusion_matrix(y_test, y_predict_k1) 
print(cmk1)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(cmk1, annot= True, fmt='g', cmap='Blues')
plt.show()
```

```python
print(classification_report(y_test, y_predict_k1))
```

# K = 3

```python
knnmodelk3 = KNeighborsClassifier(n_neighbors = 3) 
knnmodelk3.fit(X_train, y_train)
y_predict_k3= knnmodelk3.predict(X_test)
```

```python
cmk3= confusion_matrix(y_test, y_predict_k3) 
print(cmk3)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(cmk3, annot= True, fmt='g', cmap='Blues')
plt.show()
```

# K = 5

```python
knnmodelk5 = KNeighborsClassifier(n_neighbors = 5) 
knnmodelk5.fit(X_train, y_train)
y_predict_k5= knnmodelk5.predict(X_test)
```

```python
cmk5= confusion_matrix(y_test, y_predict_k5) 
print(cmk5)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(cmk5, annot= True, fmt='d', cmap='Blues')
plt.show()
```
