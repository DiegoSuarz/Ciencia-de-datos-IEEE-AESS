```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
import matplotlib.pyplot as plt

# Crear el DataFrame
data = {
    'temperature': [15, 20, 25, 10, 30, 22, 18],
    'pressure': [101.3, 102.0, 100.5, 99.8, 101.0, 100.0, 98.5],
    'humidity': [70, 65, 60, 75, 55, 67, 80],
    'rpm': [3000, 3200, 2800, 2900, 3300, 3100, 2700],
    'performance': [85, 88, 80, 82, 90, 87, 78]
}

df = pd.DataFrame(data)

# Definir las caractersticas (X) y la variable objetivo (y)
X = df[['temperature', 'pressure', 'humidity', 'rpm']]
y = df['performance']

# Crear el modelo de rbol de decisin
model = DecisionTreeRegressor()

# Entrenar el modelo
model.fit(X, y)

# Visualizar el rbol
plt.figure(figsize=(20,10))
tree.plot_tree(model, feature_names=['temperature', 'pressure', 'humidity', 'rpm'], filled=True)
plt.show()

# Realizar la prediccin
prediction = model.predict([[25, 101.5, 60, 3000]])
print(prediction)
```

El modelo predice que el motor va a tener un rendimiento de 90%
