```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Crear el DataFrame
data = {
    'sqft': [2104, 1600, 2400, 1416, 3000, 1985, 1534],
    'bedrooms': [3, 3, 3, 2, 4, 4, 3],
    'age': [10, 5, 20, 15, 8, 12, 30],
    'price': [399900, 329900, 369000, 232000, 539900, 299900, 314900]
}

df = pd.DataFrame(data)
```

```python
# Definir las características (X) y la variable objetivo (y)
X = df[['sqft', 'bedrooms', 'age']]
y = df['price']
```

```python
# Crear el modelo de árbol de decisión
model = DecisionTreeRegressor()
```

```python
# Entrenar el modelo
model.fit(X, y)
```

```python
# Visualizar el árbol
plt.figure(figsize=(20,10))
tree.plot_tree(model, feature_names=['sqft', 'bedrooms', 'age'], filled=True)
plt.show()
```

```python

# Realizar la predicción con valores especificos
prediction = model.predict([[2000, 3, 10]]) #se corresponde a las variables independintes X = df[['sqft', 'bedrooms', 'age']]
print(prediction)
```
