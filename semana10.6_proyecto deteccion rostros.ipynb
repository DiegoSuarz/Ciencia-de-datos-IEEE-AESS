## Generar data de entrada de las fotos de Elon

```python
import cv2 as cv
import os
import imutils

# Definir el nombre del directorio donde se guardarán las fotos de los rostros detectados.
modelo = 'FotosElon'
# Ruta base donde se encuentra el directorio del modelo.
ruta1 = 'archivos\\archivos opencv'
# Construir la ruta completa al directorio del modelo.
rutacompleta = ruta1 + '/' + modelo
# Crear el directorio si no existe.
if not os.path.exists(rutacompleta):
    os.makedirs(rutacompleta)

# Inicializar la captura de video desde un archivo específico.
camara = cv.VideoCapture('archivos\ElonMusk.mp4')

# Cargar el clasificador preentrenado para la detección de rostros frontales.
ruidos = cv.CascadeClassifier('opencv_complementos\haarcascade_frontalface_default.xml')


# Inicializar el contador de identificación de imágenes.
id = 0
# Bucle para leer cada frame del video.
while True:
    respuesta, captura = camara.read()
    # Si no se puede leer un frame, salir del bucle.
    if respuesta == False: break
    # Redimensionar la captura a un ancho de 640 píxeles.
    captura = imutils.resize(captura, width=640)
    # Convertir la imagen a escala de grises para la detección de rostros.
    grises = cv.cvtColor(captura, cv.COLOR_BGR2GRAY)
    # Copia de la captura en escala de grises para procesamiento.
    idcaptura = captura.copy()
    # Detectar rostros en la imagen.
    cara = ruidos.detectMultiScale(grises, 1.3, 5)

    # Iterar sobre cada rostro detectado.
    for (x, y, e1, e2) in cara:
        # Dibujar un rectángulo alrededor del rostro detectado.
        cv.rectangle(captura, (x, y), (x + e1, y + e2), (0, 255, 0), 2)
        # Extraer el rostro de la imagen.
        rostrocapturado = idcaptura[y:y + e2, x:x + e1]
        # Redimensionar el rostro capturado a 160x160 píxeles.
        rostrocapturado = cv.resize(rostrocapturado, (160, 160), interpolation=cv.INTER_CUBIC)
        # Guardar el rostro capturado en un archivo.
        cv.imwrite(rutacompleta + '/imagen_{}.jpg'.format(id), rostrocapturado)
        # Incrementar el contador de identificación.
        id = id + 1

    # Mostrar el resultado en una ventana.
    cv.imshow("Resultado rostro", captura)
    # Terminar el bucle después de capturar 350 rostros.
    if id == 350:
        break

# Liberar la cámara y cerrar todas las ventanas.
camara.release()
cv.destroyAllWindows()
```

## Entrenar la data obtenida en el paso anterior:

```python
import cv2 as cv
import os
import numpy as np
from time import time

# Ruta al directorio donde se almacenan los datos de rostros.
dataRuta = "archivos\\archivos opencv"
# Listar los subdirectorios en el directorio de datos, cada uno representa una persona diferente.
listaData = os.listdir(dataRuta)


# Inicializar listas para almacenar los identificadores y los rostros.
ids = []
rostrosData = []
id = 0

# Registrar el tiempo inicial del proceso de lectura de imágenes.
tiempoInicial = time()

# Recorrer cada subdirectorio (persona) en el directorio de datos.
for fila in listaData:
    rutacompleta = dataRuta + '/' + fila
    print('Iniciando lectura...')
    # Recorrer cada imagen en el subdirectorio.
    for archivo in os.listdir(rutacompleta):
        print('Imagenes: ', fila + '/' + archivo)
        # Añadir el ID correspondiente al identificador de la persona.
        ids.append(id)
        # Leer la imagen en escala de grises y añadirla a la lista de rostros.
        rostrosData.append(cv.imread(rutacompleta + '/'+ archivo, 0))

    # Incrementar el ID para la siguiente persona.
    id += 1

# Calcular el tiempo total utilizado en la lectura de imágenes.
tiempofinalLectura = time()
tiempoTotalLectura = tiempofinalLectura - tiempoInicial
print('Tiempo total lectura: ', tiempoTotalLectura)

# Crear un objeto de reconocedor de rostros usando EigenFace.
entrenamientoEigenFaceRecognizer = cv.face.EigenFaceRecognizer_create()
print('Iniciando el entrenamiento... espere')

# Entrenar el reconocedor con los rostros leídos y sus correspondientes identificadores.
entrenamientoEigenFaceRecognizer.train(rostrosData, np.array(ids))

# Registrar y calcular el tiempo final de entrenamiento.
TiempofinalEntrenamiento = time()
tiempoTotalEntrenamiento = TiempofinalEntrenamiento - tiempoTotalLectura
print('Tiempo entrenamiento total: ', tiempoTotalEntrenamiento)

# Guardar el modelo entrenado en un archivo XML.
entrenamientoEigenFaceRecognizer.write('opencv_complementos\complementos creados\EntrenamientoEigenFaceRecognizer.xml')
print('Entrenamiento concluido')
```

## Reconocimiento de imagenes entrenadas mediante un video

```python
import cv2 as cv
import os
import imutils

# Ruta al directorio donde se almacenan los datos de rostros.
dataRuta = 'archivos\\archivos opencv'
# Listar los nombres de los subdirectorios, que representan las diferentes personas.
listaData = os.listdir(dataRuta)

# Crear un reconocedor de rostros usando EigenFace y cargar el modelo previamente entrenado.
entrenamientoEigenFaceRecognizer = cv.face.EigenFaceRecognizer_create()
entrenamientoEigenFaceRecognizer.read('opencv_complementos\EntrenamientoEigenFaceRecognizer.xml')

# Cargar el clasificador Haar para la detección de rostros.
ruidos = cv.CascadeClassifier('opencv_complementos\haarcascade_frontalface_default.xml')

# Inicializar la captura de video desde un archivo.
camara = cv.VideoCapture('archivos\ElonMusk.mp4')

# Procesar el video frame por frame.
while True:
    respuesta, captura = camara.read()
    # Si no se puede leer un frame, finalizar el bucle.
    if not respuesta:
        break
    
    # Redimensionar la captura para estandarizar el tamaño.
    captura = imutils.resize(captura, width=640)
    # Convertir la captura a escala de grises para la detección de rostros.
    grises = cv.cvtColor(captura, cv.COLOR_BGR2GRAY)
    
    # Detectar rostros en el frame actual.
    cara = ruidos.detectMultiScale(grises, 1.3, 5)
    for (x, y, e1, e2) in cara:
        # Extraer el rostro detectado y redimensionarlo.
        rostrocapturado = grises[y:y+e2, x:x+e1]
        rostrocapturado = cv.resize(rostrocapturado, (160, 160), interpolation=cv.INTER_CUBIC)
        
        # Predecir el rostro usando el modelo entrenado.
        resultado = entrenamientoEigenFaceRecognizer.predict(rostrocapturado)
        # Mostrar el resultado de la predicción cerca del rostro detectado.
        cv.putText(captura, '{}'.format(resultado), (x, y-5), 1, 1.3, (0, 255, 0), 1, cv.LINE_AA)
        
        # Verificar el umbral de confianza de la predicción para determinar la identidad.
        if resultado[1] < 8000:
            # Si la predicción es confiable, mostrar el nombre de la persona.
            cv.putText(captura, '{}'.format(listaData[resultado[0]]), (x, y-20), 2, 1.1, (0, 255, 0), 1, cv.LINE_AA)
            cv.rectangle(captura, (x, y), (x+e1, y+e2), (255, 0, 0), 2)
        else:
            # Si la predicción no es confiable, indicar que no se encontró.
            cv.putText(captura, "No encontrado", (x, y-20), 2, 0.7, (0, 255, 0), 1, cv.LINE_AA)
            cv.rectangle(captura, (x, y), (x+e1, y+e2), (255, 0, 0), 2)
    
    # Mostrar el frame procesado.
    cv.imshow("Resultados", captura)
    # Esperar a que el usuario presione 's' para salir.
    if cv.waitKey(1) == ord('s'):
        break

# Liberar la cámara y cerrar todas las ventanas.
camara.release()
cv.destroyAllWindows()

# Se cierra con "S"
```
